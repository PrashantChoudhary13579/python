# Histogram Plot in Seaborn


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
val = sns.load_dataset("penguins")
val
```

```python
sns.displot(val["bill_depth_mm"])
plt.show()
```

```python
sns.displot(val["island"])
plt.show()
```

```python
sns.displot(val["bill_length_mm"])
plt.show()
```

```python
sns.displot(val["body_mass_g"])
plt.show()
```

```python
sns.displot(val["flipper_length_mm"],bins=[170,180,190,200,210,220,230,240],kde=True,rug=True, color = "red")
plt.show()
```

```python
sns.displot(val["flipper_length_mm"],kde=True,rug=True,color = "Green", log_scale = True)
plt.show()
```

1. Bins - It is used to make the data in a format/ range 
2. kde - It is used to plot the density in the format of a kernel density estimation curve.
3. rug - It is used to plot the density at the lower point of the plot.
4. color - to give the color of the histogram
5. log_scale - arranging the data in the format of log
