# Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x = pd.DataFrame({
    "var": var,
    "var_1": var_1
})
```

```python
# Graph in matplotlib
plt.plot(var,var_1)
```

```python
sns.lineplot(x = "var",y = "var_1", data = x)
plt.show()
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

y_1 = sns.load_dataset("penguins")
sns.lineplot(x = var,y = var_1)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm",y ="body_mass_g", data = y_1)
```

```python
sns.lineplot(x = "bill_length_mm", y = "sex", data = y_1,hue = "flipper_length_mm")
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y_1)
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y_1,hue="sex")
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y_1,hue="sex", size = 20)
plt.show()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = y_1, hue = "sex",style= "sex")
plt.show()
```

```python
sns.lineplot(x ="bill_depth_mm", y = "flipper_length_mm", data = y_1, hue = "sex",style = "sex", palette = "gist_rainbow")
plt.show()
```

```python
sns.lineplot(x = "bill_depth_mm", y = "flipper_length_mm",data = y_1, hue = "sex",style="sex",palette = "gist_rainbow",markers=["o",">"])
plt.show()
```
