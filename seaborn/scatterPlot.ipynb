# Seaborn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = var)
plt.show()
```

```python
# Separating out the data for male and female
sns.scatterplot(x ="bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex" )
plt.show()
```

```python
# Changing the style for male and female
sns.scatterplot(x ="bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex" , style = "sex")
plt.show()
```

```python
# Changing the size of male and female legend
sns.scatterplot(x ="bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex" , style = "sex", sizes=(10,100), palette = "gist_rainbow", alpha = 0.5 )
plt.show()
```

```python
# Changing the marker of male and female 
markers = {"Male":"+","Female":"s"}
sns.scatterplot(x ="bill_length_mm", y = "bill_depth_mm", data = var, hue = "sex" , style = "sex", sizes=(10,100), palette = "gist_rainbow", alpha = 1 , markers = True )
plt.show()
```
