# Seaborn Practice

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
value = sns.load_dataset("attention")
value
```

```python
sns.lineplot(x="subject",y="score" , data = value)
```

```python
sns.lineplot(x="subject",y="score" , data = value,hue ="attention" ,style="attention",markers=['o','s'] , dashes=False, legend="full",palette="Set2")
plt.grid()
plt.title("Student - A")
plt.show()
```
