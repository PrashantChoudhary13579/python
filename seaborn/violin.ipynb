# Violin Plot 


```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x = "day", y  = "total_bill", data = var, hue = "time")
plt.show()
```

```python
sns.violinplot(x = "day", y  = "total_bill", data = var, linewidth = 3,hue = "day", palette = "Dark2", saturation = 0.4)
plt.show()
```

```python
# order parameter
sns.violinplot(x = "time", y = "total_bill",data = var, order = ["Dinner", "Lunch"], color = "r")
```

```python
sns.violinplot(x = "day", y = "total_bill", hue = "sex", data = var)
```

```python
sns.violinplot(y = "day", x = "total_bill",data = var, hue = "sex", split = True, density_norm = "area", inner = "stick")
plt.show()
# inner = "quartile", "point","stick","box", None
```

```python
sns.violinplot(x = var["tip"])
plt.show()
```
