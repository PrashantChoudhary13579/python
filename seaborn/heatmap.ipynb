# HeatMap

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
var = np.linspace (1,30,30).reshape(10,3)
var
```

```python
sns.heatmap(var)
```

```python
var = sns.load_dataset("anagrams")
x = var.drop(columns = "attnr", axis = 1).head(5)
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,vmin = 0, vmax = 10, cmap = "cool", annot = True)
```

```python
ar = np.array([["a0","a1","a2","a3","a4"],
               ["b0","b1","b2","b3","b4",]])
ar
```

```python
var1 = np.linspace(1,10,10).reshape(2,5)
var1
```

```python
sns.heatmap(var1,vmin = 0, vmax = 10, cmap = "cool",annot = True)
```

```python
y = {"fontsize":20,"color":"r"}
g = sns.heatmap(var1,vmin = 0, vmax = 10, cmap = "cool",annot = ar, fmt="s", annot_kws = y, linewidth = 10, linecolor = "b", cbar = False, xticklabels = False, yticklabels = False)
g.set(title = "Learning something amazing", xlabel = "python", ylabel = "cube")
sns.set(font_scale= 1)
plt.show()
```
