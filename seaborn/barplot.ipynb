### Bar Plot in python seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x = var.island, y = var.bill_length_mm , color = "orange")
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", palette = "viridis"  )
plt.show()
```

```python
order_1 =["Biscoe","Dream","Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", palette = "viridis" , legend=False ,order= order_1 )
plt.show()
```

```python
# Changing the male and female position

order_1 =["Biscoe","Dream","Torgersen"]
sns.barplot(x = "island", y = "bill_length_mm", data = var, hue = "sex", palette = "viridis" , legend=False ,order= order_1 , hue_order = ["Female","Male"] , errorbar=("ci",100))
plt.show()
```

```python

sns.set(style = "darkgrid")

sns.barplot(x = "island", y = "bill_length_mm", data  = var, hue = "sex", hue_order = ["Female","Male"] , errorbar=('ci', 100), n_boot = 5, saturation = 100, err_kws = {'color':'yellow','linewidth':1}, capsize =0.3, dodge = True, alpha = 0.7)
plt.show()
```

```python
sns.barplot(x="bill_depth_mm", y = "bill_length_mm", data = var, orient ="h", hue = "sex", palette = 'light:skyblue')
plt.show()
```
