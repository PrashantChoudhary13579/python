# Practicing the subplot

Plotting 4 subplots in a single figure

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ((axis0,axis1),(axis2,axis3)) = plt.subplots(2,2)

x = np.arange(0,10,0.1)
y = np.cos(x)
z = np.sin(x)
u = np.tan(x)
v = np.cosh(x)

axis0.plot(x,y)
axis1.plot(x,z)
axis2.plot(x,u)
axis3.plot(x,v)

fig.tight_layout()
plt.show()
```

Now adding subtitle as well as a main Title.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ((axis0,axis1),(axis2,axis3)) = plt.subplots(2,2)

x = np.arange(0,10,0.1)
y = np.cos(x)
z = np.sin(x)
u = np.tan(x)
v = np.cosh(x)

# Adding heading
fig.suptitle("Trigonometric functions",fontsize=15)
axis0.set_title("Cosine Function")
axis1.set_title("Sine Function")
axis2.set_title("Tangent Function")
axis3.set_title("Coshine Function")

axis0.plot(x,y)
axis1.plot(x,z)
axis2.plot(x,u)
axis3.plot(x,v)

fig.tight_layout()
plt.show()
```

Now adding different colors to the graph and also changing way to make the subplots 

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axi = plt.subplots(2,2)

x = np.arange(0,10,0.1)
y = np.cos(x)
z = np.sin(x)
u = np.tan(x)
v = np.cosh(x)

# Adding heading
fig.suptitle("Trigonometric functions",fontsize=15)
axi[0,0].set_title("Cosine Function")
axi[0,1].set_title("Sine Function")
axi[1,0].set_title("Tangent Function")
axi[1,1].set_title("Coshine Function")

axi[0,0].plot(x,y,"go--")
axi[0,1].plot(x,z,"b+:")
axi[1,0].plot(x,u)
axi[1,1].plot(x,v,"p")

fig.tight_layout()

plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec

# creating the figure
fig = plt.figure()
# set height/ width of the subplot
fig.set_figheight(5)
fig.set_figwidth(5)

# creating the grid for different subplots
spec = gridspec.GridSpec(nrows = 2,ncols = 2,width_ratio = [2,1],height_ratios = [1,2],wspace = 0.5,hspace = 0.5)

fig, axi = plt.subplots(2,2)

x = np.arange(0,10,0.1)
y = np.cos(x)
z = np.sin(x)
u = np.tan(x)
v = np.cosh(x)

# Adding heading
fig.suptitle("Trigonometric functions",fontsize=15)
axi[0,0].set_title("Cosine Function")
axi[0,1].set_title("Sine Function")
axi[1,0].set_title("Tangent Function")
axi[1,1].set_title("Coshine Function")

axi[0,0].plot(x,y,"gs--")
axi[0,1].plot(x,z,"b+:")
axi[1,0].plot(x,u)
axi[1,1].plot(x,v,"p")

fig.tight_layout()

axi[0,0].set_axis_off()
axi[0,1].set_axis_off()
axi[1,0].set_axis_off()
axi[1,1].set_axis_off()

plt.show()

```
