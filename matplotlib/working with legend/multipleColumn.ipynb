# Using multiple column in legend

Example 1 :


```python
import matplotlib.pyplot as plt

plt.plot([0,3],[0,2.0],label = 'Label 1')
plt.plot([0,3],[0,2.2],label = 'Label 2')
plt.plot([0,3],[0,2.4],label = 'Label 3')
plt.plot([0,3],[0,2.6],label = 'Label 4')
plt.plot([0,3],[0,2.8],label = 'Label 5')
plt.plot([0,3],[0,3.0],label = 'Label 6')
plt.plot([0,3],[0,3.2],label = 'Label 7')

plt.legend(ncol = 3)
plt.show()
```

# Creating a Single legend for all subplots

Example 1 :


```python
import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(2,1,figsize=(17,20))

x = ['Maths','Physic','Chemistry','Physical Edu','English']

y1 = [39,25,45,34,40]
y2 = [30,45,27,40,36]

labels = ['In class 11','In class 12']

# Title for main 
fig.suptitle("Number of students who scored marks more than 80 in each subject ",fontsize = 30)

# creating a subplots
l1 = ax1.plot(x,y1,color = 'green')
l2 = ax2.plot(x,y2,color = 'red')

ax1.set_yticks(np.arange(0,51,10))
ax2.set_yticks(np.arange(0,51,10))

ax1.set_ylabel('Number of students',fontsize = 20)
ax1.set_xlabel('Subjects',fontsize = 20)
ax2.set_ylabel('Number of students',fontsize = 20)
ax2.set_xlabel('Subjects',fontsize = 20)


fig.legend([l1,l2],labels = labels,loc = "upper right")

ax1.grid(True)
ax2.grid(True)
plt.subplots_adjust(right = 0.9)
plt.show()
```

Example 2:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# plt.style.use("seaborn-v0_8-whitegrid")

fig = plt.figure()
axes = fig.subplots(2,2)

x = ['English','Hindi','Maths','Science','SST','IT']
y = ['25,45,34,46,34,24']

# 1st plot
axes[0,0].bar(x,y,color = 'green',label = 'Students passed in 2019')
axes[0,0].set_yticks(np.arange(0,51,5))

# 2nd plot
axes[0,1].bar(x,y,color = 'red',label = 'Students passed in 2020')
axes[0,1].set_yticks(np.arange(0,51,5))

for tick in axes[1,0].get_xticklabels():
    tick.set_rotation(45)
    
# 3rd plot
axes[1,0].bar(x,y,color = 'b',label = 'Students passed in 2021')
axes[1,0].set_yticks(np.arange(0,51,5))
axes[1,0].set_xlabel('Subjects',fontsize = 20)

# 4th plot
axes[1,1].bar(x,y,color = 'y',label = 'Students passed in 2022')
axes[1,1].set_yticks(np.arange(0,51,5))
axes[1,1].set_xlabel('Subjects',fontsize = 20)




lines = []
labels = []

for ax in fig.axes:
    Line, Label = ax.get_legend_handles_labels()
    # printing the label
    lines.extend(Line)
    labels.extend(Label)
plt.xticks(rotation = 45)

fig.legend(lines,labels,loc = 'upper right')
fig.tight_layout()
plt.show()
```
