# Matplotlib.pyplot.legend() in Python


1. Adding a Legend to a Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

X = [1,2,3,4,5]
Y = [1,4,9,16,25]

# Plotting the graph
plt.plot(X,Y)

# Function for adding the legend
plt.legend(['Single element'])

plt.show()
```

2. Changing the position of the Legend and adding the title

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = [2,0,4.5]
y2 = [1,1.5,3.5]

# plotting the graph of the function
plt.plot(y1)
plt.plot(y2)

# adding a legend to the figure
plt.legend(["blue","orange"],loc = "lower right")

# showing the plot
plt.show()
```

3. Combine Multiple labels in legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)

y1 = [2,1.2,4.5,8.7,3]
y2 = [1,1.5,3.5,10,2]

plt.plot(x,y1,label = "steel cost")
plt.plot(x,y2, label="Iron cost")


# Adding legend
plt.legend()

plt.show()
```

4. Plotting Sine and Cosine Functions with Legends in Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
fig,ax = plt.subplots()

y = np.cos(x)
z = np.sin(x)

ax.plot(x,y,'--b')
ax.plot(x,z, c='r')

ax.axis("equal")

plt.show()
```

Adding the label (now)


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
fig, ax = plt.subplots()

y = np.cos(x)
z = np.sin(x)

ax.plot(x,y,'g',label = 'Sine')
ax.plot(x,z,'r-',label = 'Cosine')

ax.axis("equal")
ax.legend(loc = "lower left")


plt.show()
```

Placing the legend outside the plot in Matplotlib

```python
import numpy as np 
import matplotlib.pyplot as plt

x = [0,1,2,3,4,5,6,7,8]
y1 = [0,3,6,9,12,15,18,21,24]
y2 = [0,1,2,3,4,5,6,7,8]

plt.plot(x,y1,label="y = 3x")
plt.plot(x,y2,label="y = x")

# adding the legend
plt.legend(bbox_to_anchor = (.70,1.15),ncols = 2 )
plt.show()
```
