# Matplotlib.pyplot.legend() in Python


1. Adding a Legend to a Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

X = [1,2,3,4,5]
Y = [1,4,9,16,25]

# Plotting the graph
plt.plot(X,Y)

# Function for adding the legend
plt.legend(['Single element'])

plt.show()
```

2. Changing the position of the Legend and adding the title

```python
import numpy as np
import matplotlib.pyplot as plt

y1 = [2,1.2,4.5]
y2 = [1,1.5,3.5]
plt.plot(y1,"o-")
plt.plot(y2,"s-")
# Adding title
plt.suptitle("Graph",fontsize = 30)
# Adding legend
plt.legend(['Blue','green'],loc = "lower right")
plt.show()
```

3. Combine Multiple labels in legend

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(5)

y1 = [2,1.2,4.5,8.7,3]
y2 = [1,1.5,3.5,10,2]

plt.plot(x,y1,label = "steel cost")
plt.plot(x,y2, label="Iron cost")


# Adding legend
plt.legend()

plt.show()
```
