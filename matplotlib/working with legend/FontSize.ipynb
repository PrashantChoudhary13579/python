# Changing Legend font size

Example 1: Using fontsize

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (8,4))

x = ['Prashant','Digvijay','Harsh','Aman','Aryan']
y = [6,9,4,2,5]

plt.bar(x,y,color = 'g')
plt.xlabel('Friends',fontsize = 15)
plt.ylabel('Home Tour')

# Default font size of the legend
plt.legend(['Home Tour'])
plt.show()
```

Example 2: Changing the text font size


```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))

x = ['Prashant','Digvijay','Harsh','Aman','Aryan','Akshit','Yuvraj','Mishra']
y = [4,9,5,1,3,8,7,0]

plt.bar(x,y,label= "Home Tour",color = 'skyblue')

plt.xlabel('Friends',fontsize = 18)
plt.ylabel('Home Tour',fontsize = 18)

plt.legend(fontsize = 15)

plt.show()
```

Example 3: Using prop keyword


```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (8,4))

plt.plot([1, 2, 4, 8, 30, 1])
plt.plot([1, 6, 13, 20, 38, 1])
plt.plot([1, 4, 8, 14, 20, 1])
plt.plot([1, 3, 4, 5, 10, 1])

plt.legend(['Blue','Green','orange','red'],prop = {'size': 15},loc = 'upper left',shadow =True,facecolor= "skyblue")
plt.show()
```

Using pyplot.rc Change Legend Font size 

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

x = np.linspace(0,(2*np.pi),endpoint = True)

xlim(x.min(),x.max())

xticks([0, np.pi/2,np.pi,(3*np.pi/2),(2*np.pi)],[r'$0$', r'$+\pi/2$', r'$+\pi$', r'$3/2\pi$', r'$2\pi$'])

ylim()
yticks([-1,0,+1],[r'$-1$', r'$0$', r'$+1$'])

plt.plot(x,np.sin(x), label = 'Sin(x)')
plt.plot(x,np.cos(x), label = 'Cos(x)')

plt.title("Trigonometric Functions",fontsize = 20)
plt.xlabel("Angles",fontsize = 18)
plt.ylabel("Values",fontsize = 18)

plt.legend(loc = 'upper center')
plt.rc('legend',fontsize = 10)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from pylab import *

x = np.linspace(0,(3*np.pi),endpoint = True)
xlim(x.min(),x.max())
xticks([0,np.pi,2*np.pi,3*np.pi, 4 *np.pi],[r'$0$',r'$\pi$',r'$2\pi$',r'$3\pi$',r'$4\pi$'])

ylim()
yticks([-1,0,1],[r'$-1$',r'$0$',r'$1$'])

plt.plot(x,np.sin(x),label='Sin(x)')
plt.plot(x,np.cos(x),label='Cos(x)')

plt.title("Trigonometric Functions ", fontsize = 20)

plt.xlabel('Angles',fontsize = 15)
plt.ylabel('Values',fontsize = 15)

plt.legend(loc = 'upper right')
plt.rc('legend',fontsize = 10)

plt.tight_layout()
plt.show()
```
