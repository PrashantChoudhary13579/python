# Matplotlib.axes.Axes.legend() in Python

Example 1: 

```python
import matplotlib.pyplot as plt
import numpy as np

fig,ax = plt.subplots()

line1, = ax.plot([1,2,3],label = "Line 1",color = "Black",linewidth = 4,linestyle = ':')

line2, = ax.plot([3,2,1],label = "Line 2",color = "Blue",linewidth= 10,linestyle = '--')

first_legend = ax.legend(handles = [line1],loc= 'upper center')
ax.add_artist(first_legend)

ax.legend(handles = [line2],loc= 'lower center')
fig.suptitle('matplotlib.axes.Axes.legend() function Example\n',fontsize = 10,fontweight = 'bold')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.rand(82841)

fig,ax = plt.subplots()
tab =["green","red","yellow","orange","purple","black"]
for color in tab:
    n = 10
    x,y = np.random.rand(2,n)
    ax.scatter(x,y,c=color,s=100,label= color,alpha = 0.8)

ax.legend(bbox_to_anchor = (1,1.025))
ax.grid("True")
fig.suptitle("Randomized Generated Data",fontsize = 20)
plt.show()
```

Keypoints

1. we can change the color brightness by applying alpha value which range between 0 to 1 in ax.scatter().
2. np.random.rand() - generating random values which is equal to the value inside ().
3. In this i created the dots representation which is done with the help of the ax.scatter()  if I used the ax.plot() it will plot the lines and so on....
4. ax.grid() - showing the grid lines in the figure
5. fig.suptitle() - used for adding the main title to the figure 
6. ax.legend(bbox_to_anchor = (1,1.025)) - used to make the legend placed outside the plot.
