# Changing the legend position 


Example 1 :

```python
import matplotlib.pyplot as plt
import numpy as np

# creating the data
x = np.linspace(1,20,20)
np.random.seed(1)
y = np.random.randint(0,40,20)

# Plotting the graph
plt.plot(x,y)
# adding the legend
plt.legend(["Legend"])
plt.show()
```

Example 2:


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,50,50)
np.random.seed(1)
y = np.random.randint(0,20,50)

plt.scatter(x,y)
plt.legend(['Legend'],loc = 0)
plt.show()
```

Example 3 :

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,50,50)
np.random.seed(1)
y = np.random.randint(0,20,50)

plt.bar(x,y)
plt.legend(['Legend'],loc = "upper left")
plt.show()
```

Example 4:

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1,50,50)
np.random.seed(1)
y = np.random.randint(0,20,50)

plt.plot(x,y, label = "Legend")
plt.legend(loc = "lower right" )
plt.show()
```
