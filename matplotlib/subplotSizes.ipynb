# Creating subplots of different sizes in Matplotlib

Method 1: Using Gridspec 


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import gridspec

# creating a figure
fig = plt.figure()

# now creating the height and width
fig.set_figheight(10)
fig.set_figwidth(10)

# creating grid
spec = gridspec.GridSpec(ncols = 2,nrows = 2,wspace=0.5, hspace = 0.5, width_ratios = [2,1], height_ratios = [1,3] )

# Initializing the values of x and y 
x = np.arange(0,10,0.1)
y = np.cos(x)
z = np.sin(x)
u = np.tan(x)
v = np.cosh(x)

axis0 = fig.add_subplot(spec[0])
axis0.plot(x,y)

axis1 = fig.add_subplot(spec[1])
axis1.plot(x,z)

axis2 = fig.add_subplot(spec[2])
axis2.plot(x,u)

axis3 = fig.add_subplot(spec[3])
axis3.plot(x,v)

# axis4 = fig.add_subplot(spec[4])
# axis4.plot(x,z)

plt.show()
```

Method 2: Using gridspec_kw

```python
import matplotlib.pyplot as plt
import numpy as np

# Setting different parameters to adjust each grid
fig, axs = plt.subplots(nrows= 2,ncols= 2,figsize = (7,7), gridspec_kw={
    'width_ratios':[3,3],'height_ratios':[3,3],'wspace':0.4,'hspace':0.4})

# Initializing x,y axis value
x = np.arange(0,10,0.1)
y = np.tan(x)
axs[0][0].plot(x,y)
axs[0][1].plot(x,y)
axs[1][0].plot(x,y)
axs[1][1].plot(x,y)

# fig.tight_layout()
plt.show()
```

Method 3: Using subplot2grid

```python
import matplotlib.pyplot as plt
import numpy as np

# creating figure and grid

fig = plt.figure()
fig.set_figheight(8)
fig.set_figwidth(8)

ax1 = plt.subplot2grid(shape= (3,3), loc= (0,0), colspan = 3)
ax2 = plt.subplot2grid(shape= (3,3), loc= (1,0), colspan = 2)
ax3 = plt.subplot2grid(shape= (3,3), loc= (1,2), rowspan = 2)
ax4 = plt.subplot2grid((3,3),(2,0))
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1)

# initializing the values of x,y 
x = np.arange(0,10,0.1)
y = np.cos(x)
a = np.sin(x)
b = np.tan(x)
c = np.cosh(x)
d = np.exp(x)

# Plotting the subplot
ax1.plot(x,y)
ax1.title.set_text("cosine fn")
ax2.plot(x,a)
ax2.title.set_text("sine fn")
ax3.plot(x,b)
ax3.title.set_text("tangent fn")
ax4.plot(x,c)
ax4.title.set_text("cosh fn")
ax5.plot(x,d)
ax5.title.set_text("exp fn")

plt.tight_layout()
plt.show()
```
