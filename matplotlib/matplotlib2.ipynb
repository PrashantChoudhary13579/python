# Creating multiple subplots in Matplotlib in Python

Using matplot.pyplot.subplots

1. Using 1D array of subplots

```python
import matplotlib.pyplot as plt
x = [1,2,3]
y = [0,1,0]
z = [1,0,1]

fig ,ax = plt.subplots(2)
ax[0].plot(x,y)
ax[1].plot(x,z)
```

Subplots in Matplotlib using stacking in two directions

```python
import matplotlib.pyplot as plt
import numpy as np

# Data plotting
x = np.arange(0.0,2.0,0.01)
y = 1+ np.sin(2 * np.pi * x)

# creating 6 subplots and unpacking the output 
fig , ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)

ax1.plot(x,y,color= "orange")
ax2.plot(x,y,color= "green")
ax3.plot(x,y,color= "blue")
ax4.plot(x,y,color= "red")
ax5.plot(x,y,color= "magenta")
ax6.plot(x,y,color= "black")
```

Matplotlib Multiple Plots Same Figure


```python
import matplotlib.pyplot as plt

x = [1,2,3]
y = [4,5,6]
fig, ax = plt.subplots(2,2)
ax[0,0].plot([1,2,3],y)
ax[0,1].scatter(x,y)
ax[1,0].bar(x,y)
ax[1,1].hist([1,2,2,3,3,3,3,3,4,4,5])

plt.show()
```

# Different ways to add the title to subplots in matplotlib

Example 1 : Using set_title() method

```python
import numpy as np 
import matplotlib.pyplot as plt

# create data
x = np.array([1,2,3,4,5])

# Making subplots
fix , ax = plt.subplots(2,2)

# setting the data in each subplots
ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)

# setting the title of each subplots
ax[0,0].set_title("Linear")
ax[0,1].set_title("Double")
ax[1,0].set_title("Square")
ax[1,1].set_title("Cube")

# set spacing
fix.tight_layout()
plt.show()
```

Example 2 : Using title.set_text() method

```python
import numpy as np
import matplotlib.pyplot as plt

#creating the data 
x = np.array([1,2,3,4,5])
# Making the subplots
fig, ax = plt.subplots(2,2)

# assigning the value to the ax 
ax[0,0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x*x)
ax[1,1].plot(x,x*x*x)

# writing the title to it 
ax[0,0].title.set_text("Linear")
ax[0,1].title.set_text("Double")
ax[1,0].title.set_text("Square")
ax[1,1].title.set_text("Cube")

# set spacing
fig.tight_layout()
plt.show()
```

Example 3: Using plt.gca().set_title() method

If you want to set the title of the current axes, you can use the plt.gca().set_title()


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,3,5,7,9])
# x = np.array([3,5,7,9])
# fig, ax = plt.subplots(4,5)
fig, ax = plt.subplots(3,2)

title = ["Linear","Double","Square","Cube","Multiply by 5","Multiply by 100"]
y = [x,x*2,x*x,x*x*x,x*5,x*100]

for i in range(6):
    plt.subplot(3,2,i+1)
    plt.plot(x,y[i])
    plt.gca().set_title(title[i])

fig.tight_layout()
plt.show()
```

Example 4: Using plt.gca().title.set_text() Method

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([12,23,34,45,56])

fig, ax = plt.subplots(3,2)

title =["Linear","Double","Square","Cube","Multiple of 5","Multiple of 100"]
y = [x,x*2,x*x,x*x*x,x*5,x*100]

for i in range(6):
    plt.subplot(3,2,i+1)
    plt.plot(x,y[i])
    plt.gca().title.set_text(title[i])

fig.tight_layout()
plt.show()
```
