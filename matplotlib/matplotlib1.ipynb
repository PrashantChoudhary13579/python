```python
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,1,4,9,16]

plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
x = [0,2,4,6,8]
y = [0,4,16,36,64]
fig , ax = plt.subplots()
ax.plot(x,y,marker="o",label= "Data Points")

ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
plt.show()

plt.scatter(x,y)
plt.show()
```

2. Using pie() for Pie Charts instead of plt.plot()

we are using the labels -> for name
and sizes -> for values 

```python
import matplotlib.pyplot as plt
labels = ["Python","C++","C","Java","Javascript"]
sizes = [40,25,15,15,5]

# plt.scatter(labels,sizes)
# plt.show()
# plt.plot(labels,sizes)
# plt.show()
plt.pie(sizes,labels= labels,autopct = '%1.1f%%')
plt.show()
```

3. Plotting bar charts using bar()
Use the bar() method to create a vertical or horizontal bar chart by specifying the x and y values.

```python
import matplotlib.pyplot as plt
categories = ['A','B','C','D']
values = [3,5,7,1]
plt.bar(categories, values)
plt.show()
```

4. Using scatter() for Scatter Plots.
A Scatter plot is useful for visualizing individual data points.

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [5,7,9,11,13]

plt.scatter(x,y)
plt.show()
```

ax.plot() function


```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-np.pi,np.pi,15)
c = np.cos(x)
s = np.sin(x)

ax = plt.axes([0.1,0.1,0.8,0.8])

ax1 = ax.plot(x,c,'rD--')
ax2 = ax.plot(x,s,'bH-')
# first character - shows the color
# middle/second character - shows the line style
# '.' , 'o' , '+' , 's' , 'D' , 'H'
# last character
# : <=> represent the dotted line
# -- <=> represent the dashed line
# - <=> represent the solid line

ax.legend(labels = ('Cosine Function','Sine Function'), loc= 'upper left')
# labels are the values you are showing.
# loc - supported values are 'best', 'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center
ax.set_title("Trigonometric Function")
# for heading on top
plt.show()
```
